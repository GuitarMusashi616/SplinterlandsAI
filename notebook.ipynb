```python
class DecksToDF:
    
    @staticmethod
    def to_string(deck):
        card_tuples = [deck.elo]
        for x in deck.df_cards:
            card_tuples.append(x.Card)
        while len(card_tuples) < 8:
            card_tuples.append(np.nan)

        return card_tuples  
    
    @classmethod
    def from_decks(cls, decks):
        attribs = ["ELO Rating", "Summoner", "1st Position", "2nd Position", "3rd Position", "4th Position", "5th Position", "6th Position"]
        cards = [cls.to_string(deck) for deck in decks]
        
        return pd.DataFrame(cards, columns=attribs)
            
        
        
```

```python
def battle(p1, p2, p1_score):
    assert 0 <= p1_score <= 1
    prob_p1_wins = prob_A_wins(p1, p2)
    print(f"p1 has {round(prob_p1_wins*100)}% chance of winning")
    p1 = new_rating(p1, p1_score, prob_p1_wins)
    p2 = new_rating(p2, 1-p1_score, 1-prob_p1_wins)
    return p1, p2
```
